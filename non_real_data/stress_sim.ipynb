```python
import csv
import random
from datetime import datetime, timedelta

# Einstellungen
SENSOR_ID = "Puls_sensor"
DATEINAME = "puls_stress_daten.csv"
ANZAHL_MESSUNGEN = 3600

INTERVAL_SECONDS = 60  # in 5 min Abständen


startzeit = datetime(2026, 2, 3)

# Anfangswerte
puls = random.randint(60, 75)
stress = random.randint(20, 40)

# Aktivitätsstatus
aktiv = False
aktiv_restzeit = 0

```

```python
def stress_level(score):
    if score < 35:
        return "low"
    elif score < 65:
        return "medium"
    else:
        return "high"
```

```python
with open(DATEINAME, "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["timestamp", "heart_rate_bpm", "stress_level"])

    timestamp = startzeit

    for i in range(ANZAHL_MESSUNGEN):

        # Gelegentlich Aktivitätsphase starten
        if not aktiv and random.random() < 0.002:
            aktiv = True
            aktiv_restzeit = random.randint(60, 300)

        if aktiv:
            ziel_puls = random.randint(95, 140)
            ziel_stress = random.randint(60, 95)

            aktiv_restzeit -= 1
            if aktiv_restzeit <= 0:
                aktiv = False
        else:
            ziel_puls = random.randint(58, 80)
            ziel_stress = random.randint(15, 60)

        # Langsame Annäherung an Zielwerte
        puls += (ziel_puls - puls) * 0.05
        stress += (ziel_stress - stress) * 0.03

        # Natürliche Schwankungen
        puls += random.uniform(-1.5, 1.5)
        stress += random.uniform(-2, 2)

        # Grenzen setzen
        puls = max(50, min(180, puls))
        stress = max(0, min(100, stress))

        timestamp += timedelta(seconds=INTERVAL_SECONDS)

        writer.writerow([
            timestamp.isoformat(),
            round(puls),
            stress_level(stress)
        ])

print(f"CSV-Datei '{DATEINAME}' wurde erstellt.")
```
